```python
from chatterbot import ChatBot
from chatterbot.trainers import ListTrainer
```

```python
#import time
```

```python
my_bot=Chatbot(name='Mybot',read_only=False,logic_adapters=['chatterbot.logic.MathematicalEvalation','chatterboot.logic.BestMatch'])
```

```python
small_talk=['hi there!',
            'hi',
            'how do you do?',
            'i\'m cool',
            'fine,you',
            'always cool',
            'i\'m fine',
            'i\'m happy',
            'glad to hear that',
            'i feel awesome',
            'excellent,glad to hear that',
            'not so good',
            'you are bad at this',
            'i am offended',
            'sorry to hear that',
            'what\'s your name?',
            'i\'m Mybot.ask me a math question,please.'
           ]

math_talk_1=['pythagorean theorm',
           'a squared plus b squared equals c squared.']

math_talk_2=['law of cosines',
           'c**2=a**2 + b**2 - 2* a * b * cos(gamma)']
```

```python
list_trainer=ListTrainer(Mybot)
for item in(small_talk,math_talk_1,math_talk_2):
    list_trainer.train(item)
```

```python
print(Mybot.get_response("are you feeling now?"))
```

```python
print(Mybot.get_response("do you know law of cosines?"))
```
