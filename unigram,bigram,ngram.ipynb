```python
import nltk
import string
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.corpus import stopwords
from nltk.util import ngrams
from nltk import FreqDist
```

```python
file=open('rawCorpus.txt','r')
corpus=file.read()
```

### preprocessing raw text

```python
string.punctuation
```

```python
#remove new lines and special characters
string.punctuation=string.punctuation+'“'+'”'+'’'
string.punctuation=string.punctuation.replace('.','')


file_new=""
#removing \n
for line in corpus:
    line_new=line.replace('\n',' ')
    file_new+=line_new
preprocessed_data=''.join([char for char in file_new if char not in string.punctuation])
```

```python
preprocessed_data
```

```python
sentences=sent_tokenize(preprocessed_data)
sentences[:5]
```

```python
stop_words=set(stopwords.words('english'))
```

### creating unigrams,bigrams and trigrams

```python
unigram=[]
bigram=[]
trigram=[]

for content in sentences:
    content=content.lower()
    content=word_tokenize(content)  #contain word tokens of a single sentence
    for word in content:
        if(word=='.'): #remove full stops
            content.remove(word)
        else:
            unigram.append(word)
            
    bigram.extend(ngrams(content,2))  #genertates n-grams from the sentences
    trigram.extend(ngrams(content,3))
```

```python
print(unigram[:15],'\n')
print(bigram[:15],'\n')
print(trigram[:15],'\n')
```

```python
#remove the n-grams that contain only stopwords
def stopword_removal(n,a):
    b=[]
    if n==1:  #if unigram
        for word in a:
            if word not in stop_words:
                b.append(word)
        return b
    elif n==2:  #if bigram
        for word in a:
            if word[0] not in stop_words or word[1] not in stop_words:
                b.append(word)
        return b
    else:
        for word in a:
            if word[0] not in stop_words or word[1] not in stop_words or word[2] not in stop_words:
                b.append(word)
        return b
```

```python
uni_processed=stopword_removal(1,unigram)
bi_processed=stopword_removal(2,bigram)
tri_processed=stopword_removal(3,trigram)
```

```python
print(uni_processed[:15],'\n')
print(bi_processed[:15],'\n')
print(tri_processed[:15],'\n')
```

```python
freq_bi=FreqDist(bi_processed)
freq_tri=FreqDist(tri_processed)
```
